### On which files Annotation span stretches ?

```python
import os
from pathlib import Path

file_names_list = []

path = '/home/beck/Desktop/articlenizer/data/PLoS_methods/'

file_ext = '.ann'

for file_name in os.listdir(path):
    if not file_name.endswith(file_ext): continue
    file_names_list.append(file_name) 

file_names_list.sort()

for file_name in file_names_list[:]:
    
    file_path = Path(path + file_name)
    
    with file_path.open(mode='r') as ann_file:
        
        ann_line_list = ann_file.readlines()
        
        for ann_line in ann_line_list:
            
            lbl_lst = ann_line.split('\t')
            
            #print(lbl_lst)
            
            tx_     = lbl_lst[0]
            ann_    = lbl_lst[1]
            sw_name = lbl_lst[2].strip()
            
            ann_ls = lbl_lst[1].split()
            
            if tx_.startswith('T'):
                
                an_lb  = ann_ls[0]
                beg_   = int(ann_ls[1])
                end_   = int(ann_ls[2])
            
            
                if ((tx_ == 'T29') and(sw_name == 'STRUCTURE')and (beg_ == 10369) and (an_lb =='Application_Usage_Modelling')):
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- ok")



                elif ((tx_ == 'T35') and(sw_name == 'BEAGLE')  and (an_lb =='Application_Mention')):  #  and (beg_ == 21957)
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- shifted!!")



                elif ((tx_ == 'T11') and(sw_name == 'MaxEnt') and (beg_ == 6859) and (an_lb =='Application_Mention')):
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- ok")



                elif ((tx_ == 'T1') and(sw_name == 'EthoVision') and (an_lb =='Application_Mention')):  # and (beg_ == 8393)
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- shifted!!!")


                elif ((tx_ == 'T1') and(sw_name == 'SPSS statistics') and (beg_ == 4474)and (an_lb =='Application_Usage_Analysis')):
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- ok")
                    
                if ((tx_ == 'T5') and(sw_name == 'GraphPad') and (an_lb =='Developer') and (beg_ == 9173)):
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- ok")
                    pass


                elif ((tx_ == 'T2') and(sw_name == 'Stata') and (an_lb =='Application_Usage_Analysis') and (beg_ ==2496)):
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- ok")

                elif ((tx_ == 'T3') and(sw_name == 'SPM') and (an_lb =='Application_Usage_Analysis')and (beg_ == 7482)):
                    print(f"{tx_:{10}} {sw_name:{30}} -- file name: {file_name:{15}}  -- ok")
```
